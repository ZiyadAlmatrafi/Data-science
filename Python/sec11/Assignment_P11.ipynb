```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

**Line plot**

```
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x),'-g')
plt.plot(x, np.cos(x),':r')
plt.axis('equal')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Line plot")
```

***Scatter plot***

```
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='tab:cyan');
plt.axis('equal')
```

**Bar plot**

```
x = [1,2,3,4,5]
y = [6,7,8,9,2]
x2 = [1,4,6,8,4]
y2 =[3,2,1,6,9]

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar plot")

plt.bar(x,y, label='x1', color='tab:green')
plt.bar(x2,y2, label='x2', color='tab:blue')

plt.legend()
```
